```python
import numpy as np
import scipy as sp
import sympy as sy
from matplotlib import pyplot as plt
```

```python
x1 = np.loadtxt('./data/dat1.txt')
x2 = np.loadtxt('./data/dat2.txt')
```

```python
z = np.array([x1, x2])
sigma = np.linalg.eig(np.cov(z))
D = sigma[0]
D = 1/np.sqrt(D)
E = sigma[1]
tmp = np.dot(E, np.diag(D))
V = np.dot(tmp, E.T)
z = np.dot(V,z)
size = len(x1)
```

```python
w = np.array([1,1])
wT = w.T
wT = wT/np.linalg.norm(wT)
former = w
while(1):
  wT = w.T
  tmp = np.power(np.dot(wT, z), 3)
  tmp = np.dot(z, tmp)
  tmp = tmp/size - 3*w
  tmp /= np.linalg.norm(tmp)
  w = tmp
  if(np.isclose(w,former).all() or np.isclose(w,former*(-1)).all()):
    break
  else:
    former = w
plt.plot(np.dot(w,z))

w = np.array([0.5,0.5])
wT = w.T
wT = wT/np.linalg.norm(wT)
former = w
while(1):
  wT = w.T
  tmp = np.power(np.dot(wT, z), 3)
  tmp = np.dot(z, tmp)
  tmp = tmp/size - 3*w
  tmp /= np.linalg.norm(tmp)
  w = tmp
  if(np.isclose(w,former).all() or np.isclose(w,former*(-1)).all()):
    break
  else:
    former = w
plt.plot(np.dot(w,z))
```
